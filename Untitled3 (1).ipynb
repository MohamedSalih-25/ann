```python
import keras
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from keras.models import Sequential
data=pd.read_csv('wdbc.data',header=None)
data
```

```python
X = data.iloc[:, 2:].values
y = data.iloc[:, 1].values

from sklearn.preprocessing import LabelEncoder
labelencoder_X_1 = LabelEncoder()
y = labelencoder_X_1.fit_transform(y)

```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from keras.layers import Dense, Dropout

classifier = Sequential()
```

```python
classifier.add(Dense(16,activation='relu'))
classifier.add(Dropout(0.1))
```

```python
classifier.add(Dense(16,activation='relu'))
classifier.add(Dense(1,activation='sigmoid'))
```

```python
classifier.build(input_shape=X_train.shape)
```

```python
classifier.summary()
```

```python
classifier.compile(loss="binary_crossentropy",optimizer="adam",metrics=['binary_accuracy'])
```

```python
classifier.fit(X_train, y_train, batch_size=150, epochs=250)
```

```python
test_loss,test_acc=classifier.evaluate(X_test,y_test)
```

```python
print('test accuracy:',round(test_acc,4))
```

```python

def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
   plt.imshow(cm, interpolation='nearest', cmap=cmap)
   plt.title(title)
   plt.colorbar()
   tick_marks = np.arange(len(classes))
   plt.xticks(tick_marks, classes, rotation=30)
   plt.yticks(tick_marks, classes)

   if normalize:
       cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]

   thresh = cm.max() / 2.
   for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
       plt.text(j, i, cm[i, j],
                horizontalalignment="center",
                color="white" if cm[i, j] > thresh else "black")

   plt.tight_layout()
   plt.ylabel('Actual class')
   plt.xlabel('Predicted class')
```

```python
from collections import Counter
from sklearn.metrics import confusion_matrix
import itertools


Y_pred = classifier.predict(X_test)

y_true=Y_pred >0.5
```

```python
confusion_mtx = confusion_matrix(y_true, y_test)

plot_confusion_matrix(confusion_mtx, classes = range(2))
```
